## Part A – Theoretical Foundation (Long Answers)

### 1. What is Inferential Statistics?

Inferential Statistics is a branch of statistics that uses sample data to make conclusions, predictions, or decisions about an entire population.  
Instead of collecting data from every individual in a population, a smaller sample is studied and the results are generalized to the whole population.  
It helps researchers estimate population parameters and test hypotheses.

---

**Example:**

A health researcher wants to know the average BMI of all adults in a city.  
Instead of surveying every adult, the researcher collects data from 500 adults and uses inferential statistics to estimate the average BMI of the entire population.


---

### 🔑 Key Points
- Works with **samples** to understand populations.  
- Provides tools for **estimation** and **hypothesis testing**.  
- Saves time and resources compared to studying an entire population.  




### 2. What is Hypothesis Testing and its Components?

Hypothesis Testing is a statistical method used to determine whether there is enough evidence in a sample to support a specific claim about a population.  
It helps researchers make decisions based on data rather than assumptions.

---

### 🔹 Components of Hypothesis Testing

- **Null Hypothesis (H₀):**  
  The statement that there is no significant effect or relationship.

- **Alternative Hypothesis (H₁):**  
  The statement that there is a significant effect or relationship.

- **Significance Level (α):**  
  The probability of rejecting the null hypothesis when it is actually true.  
  Commonly set at **α = 0.05**.

- **Test Statistic:**  
  A numerical value calculated from sample data (such as Z-score, T-score).

- **P-value:**  
  The probability of obtaining the observed result if the null hypothesis is true.

---

### 📖 Example

- **H₀:** Smoking has no effect on diabetes.  
- **H₁:** Smoking affects diabetes.



---

### ✨ Key Takeaway
Hypothesis testing provides a structured way to evaluate claims using data.  
It balances the risk of error with the need to make informed decisions about populations.




### 3. Explain Confidence Interval and Critical Value

---

### 📊 Confidence Interval (CI)

A **Confidence Interval** is a range of values that is likely to contain the true population parameter with a specified level of confidence.  
It provides an estimate rather than a single value.

- A **95% confidence interval** means that if the same study is repeated many times, approximately 95% of the intervals will contain the true population parameter.

**Example:**  
The average weight is estimated to be between **65 kg and 70 kg** with 95% confidence.

---

### 📏 Critical Value

A **Critical Value** is the threshold used to determine whether the null hypothesis should be rejected.  
It is obtained from statistical tables based on the significance level.

- For a **95% confidence level**, the critical Z-value is:  
  

\[
  Z = 1.96
  \]



**Example:**  
If the calculated Z-score is greater than **±1.96**, the null hypothesis is rejected.


---

### ✨ Key Takeaway
- Confidence intervals provide a **range estimate** for population parameters.  
- Critical values act as **decision thresholds** in hypothesis testing.  
- Together, they help researchers make reliable, data-driven conclusions.




### 4. Define P-value

A **P-value** is the probability of obtaining results at least as extreme as the observed results, assuming that the null hypothesis (H₀) is true.  
It measures the strength of evidence against the null hypothesis.

---

### 📏 Decision Rule
- If **P-value < 0.05** → Reject H₀  
- If **P-value ≥ 0.05** → Fail to Reject H₀  

---

### 📖 Example
If the p-value is **0.02**, it means there is only a **2% chance** that the observed result occurred randomly.  
Therefore, the null hypothesis is **rejected**.


---

### 🌟 Importance
- Helps determine whether a result is **statistically significant**.  
- Provides a **quantitative measure** of evidence against H₀.  
- Guides researchers in making **data-driven decisions**.




### 5. Differentiate Type I and Type II Errors

In hypothesis testing, errors can occur while making decisions.

---

### ❌ Type I Error (False Positive)
A **Type I Error** occurs when the null hypothesis (H₀) is rejected even though it is true.  

**Example:**  
Concluding that smoking causes diabetes when it actually does not.

---

### ⚠️ Type II Error (False Negative)
A **Type II Error** occurs when the null hypothesis (H₀) is accepted even though it is false.  

**Example:**  
Concluding that smoking does not affect diabetes when it actually does.

---

### 📊 Comparison Table

| **Type**        | **Meaning**                  |
|------------------|------------------------------|
| **Type I Error** | Rejecting a true H₀          |
| **Type II Error**| Accepting a false H₀         |


---

### ✨ Key Takeaway
- **Type I Error** → False alarm (detecting an effect that doesn’t exist).  
- **Type II Error** → Missed detection (failing to see an effect that does exist).  
- Both errors highlight the importance of balancing **significance level (α)** and **power (1-β)** in hypothesis testing.





### 6. Brief Description of Z-Test, T-Test, Chi-Square Test, and ANOVA

---

### 📐 Z-Test
A **Z-Test** is used to compare a sample mean with a known population mean when the population standard deviation is known and the sample size is large.  

**Example:**  
Checking whether the average weight of individuals differs from **70 kg**.

---

### 📊 T-Test
A **T-Test** is used to compare the means of two groups when the population standard deviation is unknown.  

**Example:**  
Comparing average weight between **males and females**.

---

### 🔢 Chi-Square Test
A **Chi-Square Test** is used to determine whether there is an association between two categorical variables.  

**Example:**  
Finding whether **smoking status** is related to **diabetes**.

---

### 📈 ANOVA (Analysis of Variance)
**ANOVA** is used to compare the means of three or more groups.  

**Example:**  
Comparing **BMI across different age groups**.

---

### 📊 Summary Table

| **Test**          | **Purpose**                                      | **Example**                                |
|--------------------|--------------------------------------------------|--------------------------------------------|
| **Z-Test**         | Compare sample mean with population mean (large n, σ known) | Average weight vs. 70 kg                   |
| **T-Test**         | Compare means of two groups (σ unknown)          | Male vs. female average weight             |
| **Chi-Square Test**| Association between categorical variables        | Smoking status vs. diabetes                |
| **ANOVA**          | Compare means of 3+ groups                       | BMI across age groups                      |

---



### ✨ Key Takeaway
- **Z-Test & T-Test** → Compare means.  
- **Chi-Square Test** → Tests relationships between categories.  
- **ANOVA** → Compares multiple group means.  
Together, these tests form the backbone of statistical inference for different types of data.




### 7. What is Covariance?

Covariance is a statistical measure that indicates the **direction of the relationship** between two variables.  
It shows whether two variables increase or decrease together.

---

### 📊 Interpretation
- **Positive Covariance** → Variables move in the same direction.  
- **Negative Covariance** → Variables move in opposite directions.  
- **Zero Covariance** → No relationship between the variables.  

---

### 📖 Example
Age and BMI may have **positive covariance** if BMI tends to increase with age.

---

### 🧮 Formula


\[
\text{Cov}(X, Y) = \frac{1}{n-1} \sum (X_i - \bar{X})(Y_i - \bar{Y})
\]



Where:  
- \(X_i, Y_i\) = individual data points  
- \(\bar{X}, \bar{Y}\) = means of X and Y  
- \(n\) = number of observations  

---

### ✨ Key Takeaway
Covariance helps identify whether two variables move together or in opposite directions,  
but it does not measure the **strength** of the relationship (that’s done by correlation).




### 8. What is Correlation?

Correlation is a statistical measure that describes the **strength and direction** of the relationship between two variables.  
It ranges from **-1 to +1**.

---

### 📊 Interpretation
- **+1 → Perfect Positive Correlation**  
- **0 → No Correlation**  
- **-1 → Perfect Negative Correlation**

---

### 📖 Example
Age and BMI may show a **positive correlation** if BMI increases as age increases.


---

### 🧮 Formula


\[
r = \frac{\text{Cov}(X, Y)}{\sigma_X \cdot \sigma_Y}
\]



Where:  
- \( \text{Cov}(X, Y) \) = covariance of X and Y  
- \( \sigma_X, \sigma_Y \) = standard deviations of X and Y  

---

### 🌟 Importance of Correlation
- Helps identify **relationships** between variables.  
- Used in **prediction and forecasting**.  
- Widely applied in **healthcare, finance, and research**.  

---

### 🔍 Real-Life Example
A researcher studies the relationship between **exercise frequency** and **BMI**.  
If individuals who exercise more tend to have lower BMI values, a **negative correlation** may exist between exercise frequency and BMI.

# Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import norm
from scipy.stats import ttest_ind
from scipy.stats import chi2_contingency
from scipy.stats import f_oneway

from statsmodels.stats.weightstats import ztest
```

## Load Dataset

```python
df = pd.read_csv("health_dataset_200.csv")

print("Dataset Loaded Successfully")
print(df.head())
```

---
# TASK 1: HYPOTHESIS FORMULATION
---

```python
print("\nHypothesis 1")
print("H0: Smoking has no effect on Diabetes")
print("H1: Smoking affects Diabetes")

print("\nHypothesis 2")
print("H0: Age and BMI are not related")
print("H1: Age and BMI are related")
```

---
# TASK 2: CONFIDENCE INTERVAL
---


```python
mean_weight = df['weight'].mean()
std_weight = df['weight'].std()
n = len(df)

confidence_interval = norm.interval(
    confidence=0.95,
    loc=mean_weight,
    scale=std_weight / np.sqrt(n)
)

print("\n95% Confidence Interval:")
print(confidence_interval)
```

---
# TASK 3: CRITICAL VALUE & P-VALUE
---

```python
critical_value = 1.96

print("\nCritical Value (95% Confidence):")
print(critical_value)
```

---
# TASK 4: ONE SAMPLE Z-TEST
---

```python
z_stat, p_value = ztest(
    df['weight'],
    value=70
)

print("\nZ-Test Result")
print("Z Statistic =", z_stat)
print("P Value =", p_value)

if p_value < 0.05:
    print("Decision: Reject H0")
else:
    print("Decision: Fail to Reject H0")
```

---
# TASK 5: CHI-SQUARE TEST
---

```python
contingency_table = pd.crosstab(
    df['smoking_status'],
    df['diabetes']
)

chi2, p, dof, expected = chi2_contingency(
    contingency_table
)

print("\nChi-Square Test")
print("Chi-Square Statistic =", chi2)
print("P Value =", p)

if p < 0.05:
    print("Decision: Reject H0")
else:
    print("Decision: Fail to Reject H0")
```

---
# TASK 6: ANOVA TEST
---

```python
group1 = df[df['age_group'] == '18-25']['bmi']
group2 = df[df['age_group'] == '26-35']['bmi']
group3 = df[df['age_group'] == '36-45']['bmi']
group4 = df[df['age_group'] == '46-60']['bmi']

f_stat, p_anova = f_oneway(
    group1,
    group2,
    group3,
    group4
)

print("\nANOVA Test")
print("F Statistic =", f_stat)
print("P Value =", p_anova)

if p_anova < 0.05:
    print("Decision: Reject H0")
else:
    print("Decision: Fail to Reject H0")
```

---
# TASK 7: COVARIANCE
---

```python
covariance = df['age'].cov(df['bmi'])

print("\nCovariance")
print(covariance)
```

---
# TASK 8: CORRELATION
---

```python
correlation = df['age'].corr(df['bmi'])

print("\nCorrelation")
print(correlation)
```
